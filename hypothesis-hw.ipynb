```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('/kaggle/input/fashion-products/fashion_products.csv')
```

```python
data = df['Price']
```

```python
n = 100
m = 500000
bootstrap_means = np.zeros(m)
for i in range (m):
    bootstrap_sample = np.random.choice(data, size = n, replace = True)
    bootstrap_means[i] = np.mean(bootstrap_sample)
mean_of_means = np.mean(bootstrap_means)
plt.hist(bootstrap_means, bins = 50, density = True, alpha = 0.5)
plt.xlabel('Mean')
plt.ylabel('Probability')
plt.title('Probability Distribution of the Mean')
plt.show()
confidence_interval = np.percentile(bootstrap_means, [5, 95])
z_critical = 1.645
population_mean = np.mean(data)
population_std = np.std(data)
margin_of_error = z_critical * (population_std /np.sqrt(n))
confidence_interval_z = (population_mean - margin_of_error, population_mean + margin_of_error)
print('Mean: ', mean_of_means)
print('Bootstrap Confidence Interval: ', confidence_interval)
print('Confidence Interval (Z Distribution): ', confidence_interval_z)
```
